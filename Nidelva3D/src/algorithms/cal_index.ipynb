# This script compares the performace of calcuating the index using the two methods
1. KDTree
2. Distance Matrix

```python
import numpy as np
from pykdtree.kdtree import KDTree
import matplotlib.pyplot as plt 
from time import time
```

```python
N = np.arange(1, 10000)

t_method1 = []
t_method2 = []

for n in N: 
    print("N = ", n, "...")
    x = np.random.randn(n)
    x_unknown_location = np.array([np.random.randn()])
    x_tree = KDTree(x)

    # Method 1, KDTree
    t1 = time()
    dist, ind1 = x_tree.query(x_unknown_location)
    t_method1.append(time() - t1)
    # %timeit x_tree.query(x_unknown_location)

    # Method 2, distance matrix
    t2 = time()
    ind2 = np.argmin(np.abs(x - x_unknown_location))
    t_method2.append(time() - t2)
    # %timeit np.argmin(np.abs(x - x_unknown_location))
```

```python
plt.plot(np.log(t_method1), label='KDTree')
plt.plot(np.log(t_method2), label='Distance matrix')
plt.legend()
plt.xlabel('Number of points')
plt.ylabel('Log time (s)')
plt.show()
```
