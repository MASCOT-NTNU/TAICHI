```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.cm import get_cmap
os.listdir(os.getcwd())
path = os.getcwd() + "/raw_ctd/"
box = pd.read_csv("box_xy.csv").to_numpy()
files = os.listdir(path)
files.sort()
file = files[-2]
filepath = path + file + "/"

files_ctd = os.listdir(filepath)
files_ctd.sort()


df = np.empty([0, 5])
for i in range(len(files_ctd)):
    f = filepath + files_ctd[i]
    df = np.append(df, pd.read_csv(f).to_numpy(), axis=0)

    
path_threshold = os.getcwd() + "/threshold/"
files_threshold = os.listdir(path_threshold)
files_threshold.sort()
file_th = files_threshold[-2]
filepath_th = path_threshold + file_th + "/"
files_th = os.listdir(filepath_th)
files_th.sort()

df_th = np.empty([0, 2])
for i in range(len(files_th)):
    f = filepath_th + files_th[i]
    df_th = np.append(df_th, pd.read_csv(f).to_numpy(), axis=0)


```
